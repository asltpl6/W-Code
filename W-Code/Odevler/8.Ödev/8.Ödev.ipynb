# Dataset ve Kolon Bilgileri
Datasetini ChatGPT tarafında oluşturdum. Burdaki bilgiler gerçeği yansıtmıyor tamamen öğrenmek amaçlı oluşturulmuş datalardır !!!
* Track ID: Şarkının benzersiz kimliği.
* Track Name: Şarkının adı.
* Artist Name: Şarkıyı seslendiren sanatçının adı.
* Album Name: Şarkının bulunduğu albümün adı.
* Release Date: Şarkının yayımlandığı tarih.
* Genre: Şarkının müzik türü (örneğin, pop, rock, hip-hop).
* Duration (ms): Şarkının süresi (milisaniye cinsinden).
* Popularity: Şarkının popülarite skoru (0 ile 100 arasında bir değer).
* Danceability: Şarkının dans edilebilirlik skoru (0.0 ile 1.0 arasında bir değer).
* Energy: Şarkının enerji skoru (0.0 ile 1.0 arasında bir değer).
* Loudness (dB): Şarkının ses seviyesi (decibel cinsinden).
* Tempo (BPM): Şarkının temposu (dakikada vuruş sayısı).
* Key: Şarkının anahtarı (müzikal ton).
* Valence: Şarkının genel ruh halini belirtir (0.0 ile 1.0 arasında bir değer).
* Acousticness: Şarkının akustik özellikleri (0.0 ile 1.0 arasında bir değer).
* Instrumentalness: Şarkının enstrümantasyon seviyesi (0.0 ile 1.0 arasında bir değer).
* Speechiness: Şarkıda konuşma varlığı (0.0 ile 1.0 arasında bir değer).

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["Artist Name"].value_counts()
```

```python
# Artist Name sütunundaki değerlerin sayısını al
artist_counts = df["Artist Name"].value_counts()

# Pasta grafiği oluştur
plt.figure(figsize=(8, 8))  # Grafik boyutunu ayarlayabilirsiniz
plt.pie(artist_counts, labels=artist_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Sanatçı İsimlerinin Dağılımı")  # Grafiğe başlık ekleyin
plt.show()
```

```python
df["Artist Name"].value_counts()
sns.countplot(y=df["Artist Name"], data=df)
plt.show()
```

```python
df.groupby("Artist Name").agg({"Popularity": "sum"})
```

```python
# Sanatçı isimlerine göre popülerlik toplamını hesapla
artist_popularity = df.groupby("Artist Name").agg({"Popularity": "sum"}).reset_index()

# Çubuk grafik oluştur
plt.figure(figsize=(10, 6))  # Grafik boyutu
sns.barplot(x="Popularity", y="Artist Name", data=artist_popularity, palette="viridis")
plt.title('Sanatçıya Göre Toplam Popülerlik')
plt.xlabel('Total Popularity')
plt.ylabel('Artist Name')
plt.show()
```

### Şarkının yılını çokış yılındanki tarih saat bilgidi olan kolondan sadace yıl bilgisi alıyoruz

```python
df["Year"] = pd.to_datetime(df["Release Date"]).dt.year
```

```python
year_popularity_mean = df.groupby("Year").agg({"Popularity": "mean"})
year_popularity_mean
```

```python
year_popularity_mean = df.groupby("Year").agg({"Popularity": "mean"}).reset_index()

# Grafik oluşturma
plt.figure(figsize=(10, 6))  # Grafik boyutunu ayarlama
sns.lineplot(x="Year", y="Popularity", data=year_popularity_mean, marker='X')  # Çizgi grafik ve işaretleyici
plt.title("Yıl ve Popülarite")  # Grafik başlığı
plt.xlabel("Year")  # X eksen etiketi
plt.ylabel("Popularity")  # Y eksen etiketi
plt.grid(True)  # Grid çizgileri ekleme
plt.show()  # Grafiği gösterme
```

```python
# Şarkıların türlerine göre yıllık dağılımı
plt.figure(figsize=(12, 8))
df.groupby(["Genre", "Year"]).size().unstack().plot(kind="bar", stacked=True)
plt.title("Müzik Türlerinin Yıllara Göre Dağılımı")
plt.xlabel("Müzik Türü")
plt.ylabel("Şarkı Sayısı")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Yıl")
plt.tight_layout()
plt.show()
```
