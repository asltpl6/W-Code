Veri Sütunları  
adm_type: Yatış türünü belirtir (örneğin, "Devirden").  
shift_from: Hastanın hangi kaynaktan devredildiğini belirtir (örneğin, "ER" Acil Servis, "Neu" Nöroloji).  
ssc: Sehat Sahulat Kartı sigortası veya KPK eyaletine ait Sağlık kartı.  
yr_nae: Muhtemelen yatış olayının yılını temsil eder.  
m_no: Yatış olayının ay numarası.  
mrn: Her hasta için benzersiz bir tıbbi kayıt numarası.  
pt_name: Hastanın adı.  
sex: Hastanın cinsiyeti ("M" erkek, "F" kadın).  
disease: Hastanın teşhisi veya durumu (örneğin, "STEMI", "ADHF").  
D.O.A: Yatış tarihi (farklı formatlarda).  
D.O.D: Taburcu olma veya ölüm tarihi (farklı formatlarda).  
status: Hastanın taburcu durumu (örneğin, "Taburcu").  
consultant: Hastadan sorumlu danışman doktorun adı.  
L.O.S: Hastanede kalış süresi, gün olarak ölçülür.

# Hastalıklar ve Tıbbi Terimler

### 1. AWMI (Ön Duvar Kalp Krizi)
Kalbin ön kısmına kan gitmediğinde oluşan kalp krizi.

### 2. IWMI (Alt Duvar Kalp Krizi)
Kalbin alt kısmına kan gitmediğinde oluşan kalp krizi.

### 3. NSTEMI (Hafif Kalp Krizi)
Kalbe giden kan azalır, ancak tamamen kesilmez.

### 4. MI (Kalp Krizi)
Kalp damarlarının tıkanmasıyla oluşan kalp krizi.

### 5. CHB (Tam Kalp Bloğu)
Kalp atışlarını düzenleyen sinyallerde ciddi bozukluk.

### 6. IHD (Kalp Damar Hastalığı)
Kalbe yeterli kan gitmez, göğüs ağrısı veya kalp krizine yol açabilir.

### 7. ADHF (Akut Kalp Yetmezliği)
Kalp, vücudun ihtiyaçlarına yeterince kan pompalayamaz.

### 8. ACS (Akut Koroner Sendrom)
Kalbe kan sağlayan damarlar aniden daralır veya tıkanır.

### 9. Chest Pain (Göğüs Ağrısı)
Kalp krizi veya göğüsle ilgili diğer problemlerle ilişkili ağrı.

### 10. PCI (Koroner Girişim)
Kalbe kan akışını düzeltmek için daralmış damarların açılması işlemi.


```python
plt.figure(figsize=(10, 5))
top_10_data["disease"].value_counts().plot(kind="bar")
plt.title("En Çok Görülen 10 Hastalık")
plt.xlabel("Hastalık")
plt.ylabel("Hastalık görülme sayısı")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
top_10_data.groupby(["disease","sex"]).size().unstack().plot(kind="bar", stacked=True)
plt.title("Cinsiyete Göre Hastalık Dağılımı")
plt.xlabel("Hastalık")
plt.ylabel("Hasta sayısı")
plt.legend(["Kadın", "Erkek"])
plt.show()
```
