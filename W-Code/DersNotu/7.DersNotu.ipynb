```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
katagori_maas = df.groupby("job_category")[["salary_in_usd"]].mean().reset_index()

plt.figure(figsize=(25,6))
plt.bar(katagori_maas["job_category"], katagori_maas["salary_in_usd"])
plt.xlabel("iler")
plt.ylabel("maa")
plt.title("Kategorilere Gre Maalar")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(25,6))
sns.barplot(x="job_category", y="salary_in_usd", data=katagori_maas)
plt.xlabel("iler")
plt.ylabel("maa")
plt.title("Kategorilere Gre Maalar")
plt.xticks(rotation=90)
```

```python
df.groupby(["work_year","job_category"])[["salary_in_usd"]].mean().reset_index()
```

```python
year_category_salary = df.groupby(["work_year","job_category"])[["salary_in_usd"]].mean().reset_index()

plt.figure(figsize=(25,6))
sns.lineplot(x="job_category", y="salary_in_usd", hue="work_year", data=year_category_salary)
plt.xlabel("Yl")
plt.ylabel("Maa")
plt.title("Yllara Gre Kategorilere Gre Maalar")
plt.xticks(rotation=45)
```

```python
# Convert 'work_year' to string
year_category_salary['work_year'] = year_category_salary['work_year'].astype(str)

# Plot again
plt.figure(figsize=(25,6))
sns.barplot(x="job_category", y="salary_in_usd", hue="work_year", data=year_category_salary)
plt.xlabel("Yl")
plt.ylabel("Maa")
plt.show()
```
